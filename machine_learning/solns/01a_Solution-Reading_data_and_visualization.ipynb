```python
# ====================================================
# import modules
# ====================================================
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import nibabel as nib
```

```python
# ====================================================
# TODO - reading and visualizing classification data from csv files
# ====================================================

# read features from the csv file
features = pd.read_csv('data/ex1_features_classification.csv')

# read labels from the txt file
labels = np.loadtxt('data/ex1_labels_classification.txt')

# understand sizes
nsamples = features.shape[0]
ndims = features.shape[1]

# visualize positive and negative samples
pos_rows = labels > 0
neg_rows = labels <= 0
plt.plot(features.feature1[pos_rows],features.feature2[pos_rows],'+',markersize=10,mew=2)
plt.plot(features.feature1[neg_rows],features.feature2[neg_rows],'_',markersize=10,mew=2)
plt.grid('on')
plt.xlabel('feature1',fontsize=16), plt.ylabel('feature2',fontsize=16)
plt.show()
```

```python
# ====================================================
# TODO - reading and visualizing MRI image
# ====================================================

# read the .nii file
I = nib.load('../data/ex1_mri.nii.gz')

# extract the pixel values
V = I.get_data()

# visualize slices in different orientations
plt.figure(figsize = (12, 6))

cx = np.int(V.shape[0]/2) # index of the central slice along the x-direction
cy = np.int(V.shape[1]/2) # index of the central slice along the y-direction
cz = np.int(V.shape[2]/2) # index of the central slice along the z-direction

plt.subplot(1,3,1), plt.imshow(np.rot90(V[:,:,cz], k=3), cmap='gray')
plt.subplot(1,3,2), plt.imshow(np.rot90(V[:,cy,:], k=1), cmap='gray')
plt.subplot(1,3,3), plt.imshow(np.rot90(V[cx,:,:], k=1), cmap='gray')
plt.show()
```
